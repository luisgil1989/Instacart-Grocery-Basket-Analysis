```python
# Create a bar chart for the 'orders_day_of_week' column
bar = df['order_dow'].value_counts().sort_index().plot.bar(color=['purple', 'red', 'pink', 'orange', 'yellow', 'green', 'blue'])
plt.title('Orders by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Number of Orders')
plt.show()
```

```python
# Create a line chart for age and number of dependents
line_age_dependents = sns.lineplot(data=df, x='age', y='num_dependents')
plt.title('Age and Number of Dependents')
plt.xlabel('Age')
plt.ylabel('Number of Dependents')
plt.show()
```

### Age and Number of Dependents

The line chart above shows the relationship between the age of customers and the number of dependents they have. Each point on the line represents the average number of dependents for customers of a specific age.

From the chart, it can be observed that there is no consistent trend in the number of dependents as the age of customers varies. The number of dependents fluctuates across different ages, indicating that age may not be a strong predictor of the number of dependents. There are peaks and troughs throughout the chart, suggesting that other factors might be influencing the number of dependents in addition to age.

Note: The fluctuations could also be due to the averaging effect on a potentially small dataset, resulting in visible variations.


```python
# Export the scatterplot
line_age_dependents.figure.savefig(os.path.join(path_viz, 'line_age_dependents.png'))
```

```python
# Create a scatterplot for age and income
scatter_age_income = sns.scatterplot(x='age', y='income', data=df)
plt.title('Age and Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

### Age and Income

The scatterplot above illustrates the relationship between the age of customers and their income. Each point on the scatterplot represents an individual customer's age and their corresponding income.

From the chart, it can be observed that:
- Income levels appear to be widely distributed across all ages, indicating that age is not a strong predictor of income.
- There is a higher concentration of customers with incomes below $200000 across  all  age groups.
- While there are some customers with incomes above $200000 these instances are relatively sparse and scattered throughout the age range.
- There are no obvious trends or patterns suggesting a direct correlation between age and income, as customers of similar ages have varying income levels.

This scatterplot highlights the variability and diversity in income levels among customers of different ages, suggesting that other factors may play a more significant role in determining income.


```python
# Export the scatterplot
scatter_age_income.figure.savefig(os.path.join(path_viz, 'scatter_age_income.png'))
```
